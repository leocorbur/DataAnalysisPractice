```python
# Actulizando tipos de datos
# date: date - anio-mes-dia
# vru_entry, vru_exit: time
# vru_time: segundos (vru_time= exit_time – entry_time)
# q_start, q_exit: time
# q_time: segundos (q_time = q_exit – q_start)
# ser_start, ser_exit: time
# ser_time: segundos (ser_time = ser_exit – ser_start)
```

```python
# Convertir date a tipo date (asegúrate de que date esté en formato 'anio-mes-dia')
df['date'] = pd.to_datetime(df['date'])
```

```python
import time
```

```python
# Convertir vru_entry y vru_exit a tipo time
df['vru_entry'] = pd.to_datetime(df['vru_entry'], format='%H:%M:%S').dt.time
df['vru_exit'] = pd.to_datetime(df['vru_exit'], format='%H:%M:%S').dt.time
```

```python
# Calcular vru_time en segundos (vru_time = exit_time – entry_time)
df['vru_time'] = (pd.to_datetime(df['vru_exit'], format='%H:%M:%S') - pd.to_datetime(df['vru_entry'], format='%H:%M:%S')).dt.total_seconds()
```

```python
# Convertir q_start y q_exit a tipo time
df['q_start'] = pd.to_datetime(df['q_start'], format='%H:%M:%S').dt.time
df['q_exit'] = pd.to_datetime(df['q_exit'], format='%H:%M:%S').dt.time

# Calcular q_time en segundos (q_time = q_exit – q_start)
df['q_time'] = (pd.to_datetime(df['q_exit'], format='%H:%M:%S') - pd.to_datetime(df['q_start'], format='%H:%M:%S')).dt.total_seconds()
```

```python
# Convertir ser_start y ser_exit a tipo time
df['ser_start'] = pd.to_datetime(df['ser_start'], format='%H:%M:%S').dt.time
df['ser_exit'] = pd.to_datetime(df['ser_exit'], format='%H:%M:%S').dt.time

# Calcular ser_time en segundos (ser_time = ser_exit – ser_start)
df['ser_time'] = (pd.to_datetime(df['ser_exit'], format='%H:%M:%S') - pd.to_datetime(df['ser_start'], format='%H:%M:%S')).dt.total_seconds()
```
